## Line Plot

```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from numpy.random import randn, randint, uniform, sample
```

```python
df = pd.DataFrame(randn(1000), index=pd.date_range('2019-06-07', periods=1000), columns=['values'])
```

```python
df.head()
```

```python
ts = pd.Series(randn(1000), index=pd.date_range('2019-06-07', periods=1000))
```

```python
ts.head()
```

```python
df['values'] = df['values'].cumsum()
df.head()
```

```python
ts = ts.cumsum()
ts.head()
```

## Scatter Matrix and Subplots

```python
from pandas.plotting import scatter_matrix
```

```python
df
```

```python
scatter_matrix(df, figsize=(8,8), diagonal='kde', color='r')
plt.show()
```

```python
df.plot(subplots=True)
plt.show()
```

```python
df.plot(subplots=True, sharex=False)
plt.tight_layout()
plt.show()
```

```python
df.plot(subplots=True, sharex=False, layout=(2, 2), figsize=(6, 6))
plt.tight_layout()
plt.show()
```
